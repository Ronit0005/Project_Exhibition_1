```python
import pandas as pd
```

```python
from pandas.plotting import scatter_matrix
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
Data = pd.read_csv("main_set.csv")
```

Distribution Of Patient Age ( Where Curve Line Is Density Function

```python
plt.figure(figsize=(10, 6))
sns.histplot(Data['Age'], bins=10, kde=True)
plt.title('Distribution of Patient Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis="y")
plt.show()
```

```python
corr_matrix = Data.corr(numeric_only=True)
```

```python
print(corr_matrix)
```

```python
corr_matrix = Data.corr(numeric_only=True)
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
print(corr_matrix["Heart Attack Risk"].sort_values(ascending=False))
```

Here We Can See That Heart Attack Is Highly Correlated With General Health

We Often Hear That Male Is Risked More To Heart Attack Than The Women , Let Us Prove This With A Comparision of Risk Of Heart Attack In Men And Women

```python
sns.boxplot(x="Heart Attack Risk", y="Cholesterol", data=Data)
plt.title('Boxplot of Heart Attack Risk VS Colestrol Levels')
plt.grid(axis="y",alpha=0.2)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Heart Attack Risk', data=Data)
plt.title('Heart Attack Risk by Gender', fontsize=16)
plt.xlabel('Sex',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.legend(title='Heart Attack Risk',labels=['No','Yes'])
plt.show()
```

Box Plot Comparing The Cholestrol Levels To The Diet

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Diet', y='Cholesterol', data=Data)
plt.title('Cholesterol Levels by Diet')
plt.xlabel('Diet')
plt.ylabel('Cholesterol')
plt.show()
```

Scatter Matrix For Heart Attack Risk , Cholestrol , Diabetes , Exercise Hours Per Week

```python
attributes = ["Heart Attack Risk", "Cholesterol", "Diabetes", "Exercise Hours Per Week"]
scatter_matrix(Data[attributes], figsize=(12, 8))
plt.show()
```
